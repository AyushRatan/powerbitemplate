```python
main_df = pd.read_csv('/content/POC_Revenue_Opportunities_MasterData.csv', encoding="ISO-8859-1")
main_df.head()
```

```python
combined = ['Report Name: {0}; Page Name: {1}; Title: {2}; Type: {3}; Columns: {4}; Description: {5};'.format(row['Report_Name'],
                                                                                                              row['Page_Name'],
                                                                                                              row['Title'],
                                                                                                              row['Type'],
                                                                                                              row['Columns'],
                                                                                                              row['Description']) for ix, row in main_df.iterrows()]
main_df['combined_text'] = combined
main_df.head()
```

```python
def get_embeddings(text, model="text-embedding-ada-002"):
  res = openai.embeddings.create(
      input = text,
      model = model
  )

  return res.data[0].embedding

main_df['embeddings'] = main_df['combined_text'].apply(lambda x: get_embeddings(x, model = "text-embedding-ada-002"))
main_df.head()
```

```python
query = "Revenue report"
query_vector = get_embeddings(query)
```

```python
similarity_search = [main_df['embeddings'].apply(lambda x: np.dot(x, query_vector))]
```
