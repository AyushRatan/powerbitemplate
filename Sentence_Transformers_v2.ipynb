```python
#pip install sentence_transformers
import sentence_transformers
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

```python
data = pd.read_excel("Merged_MasterData.xlsx")
```

```python
# Load the pre-trained SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def calculate_cosine_similarity(query, description):
    query_embedding = model.encode(query, convert_to_tensor=True)
    description_embedding = model.encode(description, convert_to_tensor=True)
    cosine_similarity = util.cos_sim(query_embedding, description_embedding)

    return cosine_similarity.item()
```

```python
def find_top_3_matches(query):
    cosine_similarities = []
    for index, row in data.iterrows():
        description = row['Responses']
        similarity = calculate_cosine_similarity(query, description)
        cosine_similarities.append((index, similarity))

    cosine_similarities.sort(key=lambda x: x[1], reverse=True)

    top_3_matches = []
    for index, similarity in cosine_similarities[:3]:
        match = {
            'Report_Name': data.loc[index, 'Report_Name'],
            'Page_Name': data.loc[index, 'Page_Name'],
            'Visual_Title': data.loc[index, 'Visual_Title'],
            'Visual_Type': data.loc[index, 'Visual_Type'],
            'Columns': data.loc[index, 'Columns'],
            'Description': data.loc[index, 'Responses'],
            'Similarity': similarity
        }
        top_3_matches.append(match)

    return top_3_matches
```

```python
#1) Including Type
#2) Two components
#3) MVR
```

```python
query = "Give me the month wise revenue count"
top_3_matches = find_top_3_matches(query)
for match in top_3_matches:
    print(match)
```

```python
query_2 = "Give me the Revenue details in a pivot table"
top_3_matches = find_top_3_matches(query_2)
for match in top_3_matches:
    print(match)
```

```python
query_3 = "Give me the Forecast by location and opportunity count by region"
top_3_matches = find_top_3_matches(query_3)
for match in top_3_matches:
    print(match)
```
